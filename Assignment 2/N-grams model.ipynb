```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read

```python
with open('swami_vivekananda_prana.txt') as f:
    lines = f.readlines()
```

```python
lines
```

```python
no_n = list(filter(("\n").__ne__, lines))
no_n
```

```python
corpus = ""
for n in no_n:
    corpus += re.sub(r'[^\w\s]', '', re.sub(r"\n", "", n))
corpus = corpus.lower()
corpus
```

```python
broken_corpus = re.findall(r"[\w']+|[.,!?;]", corpus)

stiched_corpus = " ".join(broken_corpus)
stiched_corpus
```

```python
all_words = list(set(broken_corpus))
all_words
```

```python
len(all_words)  # Unique words
```

## Functions

```python
def occurances(sequence):
    return len(re.findall(sequence, stiched_corpus))
```

```python
def highest_occ(sequence):
    counts = []
    for i in all_words:
        inter_seq = sequence + " " + i
        counts.append(occurances(inter_seq))
    return all_words[np.argmax(counts)]
```

```python
def gram_model(n):
    max_length = 11
    n_words = []
    for _ in range(n-1):
        n_words.append(input("Input your word:  ").strip())

    possible = True
    for i in n_words:
        if not i in all_words:
            possible = False

    if possible:
        sentence = " ".join(n_words)
        for _ in range(max_length):
            input_sequence = " " + " ".join(n_words)

            new_word = highest_occ(input_sequence)
            sentence = sentence + " " + new_word
            
            n_words = n_words[1: ]
            n_words.append(new_word)
        print("The generated sentence is \n")
        print(sentence)
    else:
        print("Not Possible to generate. Not in word list.")
```

## 1- Gram

```python
gram_model(1)
```
